## Top 10 cultivos by sembrada area

### Subtask:
Calculate the total 'area sembrada' for each 'cultivo', identify the top 10, and visualize the results.


**Reasoning**:
Calculate the total 'area sembrada' for each 'cultivo', identify the top 10, and visualize the results as a bar chart.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify the columns related to 'area sembrada' and sum them, treating NaN as 0
df_concatenated['total_area_sembrada'] = df_concatenated['area sembrada_x000D_\n(ha)'].fillna(0) + df_concatenated['area sembrada (ha)'].fillna(0)

# Group by 'cultivo' and calculate the sum of the total 'area sembrada'
cultivo_area = df_concatenated.groupby('cultivo')['total_area_sembrada'].sum()

# Sort the results and select the top 10
top_10_cultivos = cultivo_area.sort_values(ascending=False).head(10)

# Create a bar chart with a modern style and green color
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cultivos.index, y=top_10_cultivos.values, color='green')

# Label the axes and add a title
plt.xlabel('Cultivo')
plt.ylabel('rea Sembrada Total (ha)')
plt.title('Top 10 Cultivos por rea Sembrada (2006-2024)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```
